KNN

```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import StandardScaler

stu = pd.read_csv('/content/student_exam_data.csv')

stu.shape

feature_columns = ['Study Hours', 'Previous Exam Score']
X = stu[feature_columns].values
y = stu['Pass/Fail'].values
print(y)




X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

classifier = KNeighborsClassifier(n_neighbors=3)

# Fitting the model
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)
print(y_pred)

cm = confusion_matrix(y_test, y_pred)
print(cm)

accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of our model is equal ' + str(round(accuracy, 2)) + ' %.')
```

Decision Tree

```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.model_selection import train_test_split

data = pd.read_csv('/content/student_exam_data.csv')

feature_columns = ['Study Hours', 'Previous Exam Score']
X = data[feature_columns].values
y = data['Pass/Fail'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, shuffle=True, random_state=42)

# Decision Tree model
dt = DecisionTreeClassifier(criterion="entropy", max_depth=6, min_samples_split=45)

# Training
dt.fit(X_train, y_train)

# Prediction
y_pred = dt.predict(X_test)
print(y_pred)

# Confusion Matrix
print(confusion_matrix(y_test, y_pred))

# Accuracy
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of the model is', round(accuracy, 2), '%')
```

Support Vector Machine

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score

data = pd.read_csv('/content/student_exam_data.csv')

feature_columns = ['Study Hours', 'Previous Exam Score']
X = data[feature_columns].values
y = data['Pass/Fail'].values

# Split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, shuffle=True, random_state=42)

# Scaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)


svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)
print(svm.predict([[6.5,80.0]]))

# Predict
y_pred = svm.predict(X_test)

# Results
print(confusion_matrix(y_test, y_pred))
accuracy = accuracy_score(y_test, y_pred)*100
print('Accuracy of SVM is', round(accuracy, 2), '%')
```
